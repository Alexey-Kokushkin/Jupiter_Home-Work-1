```python
import matplotlib.pyplot as plt
```

```python
from sympy import *
```

```python
x=Symbol('x', real=True)
```

```python
a=1
b=-6
c=8
f=a*x**2 + b*abs(x) + c
f
```

```python
# 1. Определить корни:
roots=solve(f,x)
roots
```

```python
derivate=diff(f, x)
derivate
```

```python
solve(derivate, x)  
```

```python
#  2-3. Интервалы, на которых функция возрастает и убывает:
extremum=[]
#count=0
for i in range(-10, 10):
    if derivate.subs(x, i) == 0:
        extremum.append(i)
        #count+=1
print(extremum)
#print(count)
```

```python
if is_decreasing(f, Interval(-oo, extremum[0])):
    print (f'Function decrease in {Interval(-oo,extremum[0])}')
else:
    print (f'Function increase in {Interval(-oo,extremum[0])}')
if is_decreasing(f, Interval(extremum[0], extremum[1])):
    print (f'Function decrease in {Interval(extremum[0], extremum[1])}')
else:
    print (f'Function increase in {Interval(extremum[0], extremum[1])}')
if is_decreasing(f, Interval(extremum[1], extremum[2])):
    print (f'Function decrease in {Interval(extremum[1], extremum[2])}')
else:
    print (f'Function increase in {Interval(extremum[1], extremum[2])}')
if is_decreasing(f, Interval(extremum[2], +oo)):
    print (f'Function decrease in {Interval(extremum[2], +oo)}')
else:
    print (f'Function increase in {Interval(extremum[2], +oo)}')
```

```python
# Вычислить вершину**:
function_peak=[]
item=0
while item < len(extremum):
    y=[extremum[item],f.subs(x, extremum[item])]
    item+=1
    function_peak.append(y)
print(f'Вершины функции: {function_peak}')
      
```

```python
# Определить промежутки, на котором f > 0
# Определить промежутки, на котором f < 0
roots=solve(f,x)
root1=roots.pop()
root2=roots.pop()
root3=roots.pop()
root4=roots.pop()
middle1=(root3+root4)/2
middle2=(root3+root2)/2
middle3=(root2+root1)/2
if f.subs(x,root4-0.01)>0:
    print(f'Function > 0 in {Interval(-oo,root4)}')
else:
    print(f'Function < 0 in {Interval(-oo,root4)}')
if f.subs(x,middle1)>0:
    print(f'Function > 0 in {Interval(root4, root3)}')
else:
    print(f'Function < 0 in {Interval(root4, root3)}')
if f.subs(x,(root2+root3)/2)>0:
    print(f'Function > 0 in {Interval(root3, root2)}')
else:
    print(f'Function < 0 in {Interval(root3, root2)}')
if f.subs(x,(root1+root2)/2)>0:
    print(f'Function > 0 in {Interval(root2, root1)}')
else:
    print(f'Function < 0 in {Interval(root2, root1)}')
if f.subs(x, root1+0.01)>0:
    print(f'Function > 0 in {Interval(root1, oo)}')
else:
    print(f'Function < 0 in {Interval(root1, oo)}')
    
```
